```python
import seaborn as sns
import numpy as np
```

# Data

```python
data = np.array([
    #a   p  u  h  m  s
    [.3, 0, 0, 0, 0, 0],  # arabic 
    [.4, .5, 0, 0, 0, 0], # persian
    [.1, .2, .3, 0, 0, 0],  # urdu
    [.4, .5, .9, .8, 0, 0], # hindi
    [.1, .2, .3, .3, .7, 0], # marathii 
    [.4, .5, .6, .1, .5, .8]],# sanskrit
    np.float16)
```

```python
mask = np.zeros_like(data)
mask[np.triu_indices_from(data)] = True
xl = ['Arabic', 'Persian', 'Urdu', 'Hindi', 'Marathi', 'Sanskrit'] # labels for x-axis
yl = ['Arabic', 'Persian', 'Urdu', 'Hindi', 'Marathi', 'Sanskrit'] # labels for y-axis

with sns.axes_style("white"):
    ax = sns.heatmap(data, mask=mask, annot=True, xticklabels=xl, yticklabels=yl, vmax=1, square=True, cmap=sns.cubehelix_palette(10))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    fig = ax.get_figure()
    fig.savefig('output.eps', format='eps', dpi=300, bbox_inches='tight')
    fig.savefig('output.png', dpi=300, bbox_inches='tight')
```
